# Boston DataSet


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
tf.__version__
```

```python
from tensorflow import keras
```

```python
#Import Boston dataset from SKlearn

from sklearn.datasets import load_boston
featureData, priceData = load_boston(return_X_y=True)
```

```python
print(featureData.shape)
print(priceData.shape)
```

```python
pd.DataFrame(featureData)
```

```python
pd.DataFrame(priceData)
```

```python
# Split data into Train & test

from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(featureData, priceData, test_size=.2, random_state=9)
```

```python
#Build Deep Learning MultiLayer Perceptron Model - Sequential

dnn = tf.keras.models.Sequential()

#Add first hidden layer
dnn.add(tf.keras.layers.Dense(units=7, activation='relu'))

#Add second hidden layer
dnn.add(tf.keras.layers.Dense(units=6, activation='relu'))

#Output layer
dnn.add(tf.keras.layers.Dense(units=1))
```

```python
#Compiling the Model with Optimizer

dnn.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#Train the model
dnn.fit(xtrain, ytrain, batch_size=20, epochs=500)
```

```python
#Predict the price
y_predict = dnn.predict(xtest)
```

```python
#Check the model Accuracy

from sklearn.metrics import r2_score
r2_score(ytest, y_predict)
```
