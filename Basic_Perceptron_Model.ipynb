```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

x, y = make_classification(n_samples=150, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=35,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter', s=150)
```

```python
def perceptron(x,y):

    x = np.insert(x,0,1,axis=1)
    weights = np.ones(x.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,150)
        y_hat = step(np.dot(x[j],weights))
        weights = weights + lr*(y[j]-y_hat)*x[j]
        print("weights = ", weights)
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,150)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=150)
plt.ylim(-3,2)
```
